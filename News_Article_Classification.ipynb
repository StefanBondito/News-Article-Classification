# 2. Cleaning

## Cleaning Data

Data yang tadi sudah ada di dalam dataframe akan memasuki proses pembersihan atau cleaning, dimana kita akan mengubah data menjadi huruf kecil semua, mengubah angka menjadi kosong, mengubah karakter yang bukan tergolong huruf menjadi kosong, mengganti whitespace menjadi spasi atau kosong, dan mengganti underscore menjadi spasi kosong. Hasil dari function ini nanti akan dimasukkan menjadi kolom baru untuk dataframe final_df.

```python
## Function Cleaning Dasar
def cleaning(dataframe):
    df_clean = dataframe.str.lower()
    df_clean = [re.sub(r"\d+","",i )for i in df_clean]
    df_clean = [re.sub(r'[^\w]', ' ', i)for i in df_clean]
    df_clean = [re.sub(r'_',' ',i)for i in df_clean]
    df_clean = [re.sub(r'\s+',' ',i)for i in df_clean]

    return df_clean
```

```python
clean_text = cleaning(final_df['Text'])
clean_text
```

```python
final_df.insert(1, 'Clean Text', clean_text)
```

## Tokenization

Tokenization adalah proses memecah kalimat yang ada menjadi pecahan kecil berupa kata-kata, frasa, atau sebuah elemen yang bisa digunakan sebagai token. Token ini adalah representasi satu buah informasi.

Ini perlu dilakukan karena pemecahan kalimat menjadi kata-kata mempermudah metode-metode selanjutnya, seperti filtering atau stopwords removal, dan classification. Salah satu cara yang bisa dilakukan untuk tokenization adalah dengan menggunakan function **word_tokenize** dari library **nltk.tokenize**

### Train Tokenization

```python
## Define a function for tokenization
def tokenizing(text):
  return word_tokenize(text)

train_df['tokenized'] = train_df['Clean Text'].apply(tokenizing)
```

```python
train_df
```

### Test Tokenization

```python
test_df['tokenized'] = test_df['Clean Text'].apply(tokenizing)
```

## Filtering

Filtering atau proses stopword removal adalah tahap text preprocessing yang digunakan untuk menghilangkan "stopwords" atau kata-kata yang umum digunakan sampai ditandai sebagai tidak berisi informasi penting. Kata-kata ini biasanya dianggap hanya sebagai pelengkap atau pendukung untuk kata yang memang mempunyai informasi penting. Misalnya di bahasa Inggris, sering ditemukan kata-kata seperti "the", "is", "in", atau yang lainnya.

Stopwords removal dilakukan karena membantu dalam pengurangan noise atau menghilangkan data tidak penting agar lebih mudah kita menemukan informasi penting yang dibutuhkan. Selain itu, stopwords removal juga mengurangi dimensi pada fitur kita, sehingga membuat model bisa fokus kepada informasi atau kata-kata penting.

### Train Filtering

```python
## Filtering done by removing stopwords

from nltk.corpus import stopwords
nltk.download('stopwords')

stopwords_list = set(stopwords.words('english'))

## Function to Remove Stopwords
def stopword_removal(tokenized):
  return [word for word in tokenized if word.lower() not in stopwords_list]

train_df['filtered'] = train_df['tokenized'].apply(stopword_removal)
```

```python
train_df.drop(columns=['Media'], inplace = True)
```

### Test Filtering

```python
test_df['filtered'] = test_df['tokenized'].apply(stopword_removal)
```

## Stemming or Lemmatization

Stemming and Lemmatization adalah teknik normalisasi kata-kata di text mining untuk mengubah suatu kata menjadi bentuk dasar mereka. Stemming dan Lemmatization mempunyai tujuan sama, tetapi hanya berbeda pada metode perubahan mereka. Stemming hanya mengubahnya dengan memotong imbuhan tambahan dari kata, sedangkan Lemmatization mengubahnya ke bentuk dasar dengan mempertimbangkan bentuk pada kamus kata yang digunakan.

Untuk kasus seperti ini, saya menggunakan Lemmatization karena data berasal dari berita-berita dimana konteks setiap katanya penting tergantung pemakaiannya dan dimana dipakainya. Tujuan penggunaan metode ini adalah untuk normalisasi yang membantu mengurangi jumlah fitur agar model bisa bekerja lebih efisien.

### Train Lemmatization

```python
## Use Lemmatization because we want a more accurate transformation of the words
nltk.download('wordnet')
from nltk import WordNetLemmatizer
lemma = WordNetLemmatizer()

def lemmatizer(tokens):
  return[lemma.lemmatize(word) for word in tokens]

train_df['lemmatized'] = train_df['filtered'].apply(lemmatizer)
```

## TF-IDF

TF-IDF adalah teknik dimana kata-kata diberikan sebuah weight berdasarkan frekuensi atau keseringan munculnya kata tersebut dalam data yang nantinya dibandingkan dengan seluruh korpus dokumen atau semua dokumen. ## Word2Vec - Skipgram

**Word2Vec** adalah salah satu metode penyisipan atau penyematan kata yang sering ditemukan di text mining. Cara kerja dari Word2Vec adalah kata-kata sudah digambarkan menjadi satu buah vektor dan vektor ini nantinya akan dipelajari secara neural network. Vektor akan berusaha untuk menangkap karakteristik dari kata dan akan dilihat hubungannya berdasarkan arti, konteks penggunaan, dan lainnya. Hasil ini akan direpresentasikan dalam bentuk angka yang nantinya bisa diproses oleh model.

Sumber:

https://towardsdatascience.com/word2vec-explained-49c52b4ccb71


```python
import gensim
from gensim.models import Word2Vec
```

Word2Vec mempunyai dua arsitektur utama, yaitu Continuous Bag of Words (CBOW) dan Skipgram. Untuk kasus ini, saya menggunakan Skipgram, karena tujuan dari skipgram adalah memprediksi context sebuah kata atau data berdasarkan input dalam bentuk kata. Hal ini sangat sesuai dengan kasus ini, dimana kita mempunyai input atau X dalam bentuk kata-kata dari artikel berita dimana kita akan mencoba menggunakan kata-kata tersebut untuk menentukan pengelompokkan berita tersebut ke dalam labelnya.

Model Word2Vec bisa dibuat dengan bantuan library **gensim**. Pembuatan model akan didasarkan pada data training, dimana hasil model akan digunakan untuk membuat sebuah korpus atau kamus yang dipakai untuk memproses data test nantinya.

```python
## Create Skipgram Model
skipgram_model = gensim.models.Word2Vec(train_df['tokenized'], min_count = 3, vector_size = 50, window = 5, sg=1)
```

```python
## Making the vocabulary for skipgram from train data
vocabulary_skipgram = skipgram_model.wv.index_to_key
print(vocabulary_skipgram)
```

```python
## Mean vectors
def calculate_mean_vectors(tokens):
  token_vectors = [skipgram_model.wv[token] for token in tokens if token in vocabulary_skipgram]
  if token_vectors:
    return np.mean(token_vectors, axis = 0)
  else:
    return [0]*skipgram_model.vector_size

## Skipgram data frames
skipgram_train = pd.DataFrame()
skipgram_test = pd.DataFrame()

skipgram_train['word_vectors'] = train_df['tokenized'].apply(calculate_mean_vectors)
skipgram_test['word_vectors'] = test_df['tokenized'].apply(calculate_mean_vectors)

columns = [f'{i+1}' for i in range(skipgram_model.vector_size)]

skipgram_train[columns] = pd.DataFrame(skipgram_train['word_vectors'].tolist(), index=skipgram_train.index)
skipgram_test[columns] = pd.DataFrame(skipgram_test['word_vectors'].tolist(), index=skipgram_test.index)

skipgram_train.drop(columns=['word_vectors'], inplace=True)
skipgram_test.drop(columns=['word_vectors'], inplace=True)
```

## TFIDF Models

SVC adalah sebuah model Machine Learning yang digunakan untuk memisahkan data berdasarkan sebuah hyperplane. Hal ini bertujuan agar data-data dengan kategori atau kelas berbeda terpisahkan oleh sebuah celah yang sebesar mungkin.

Untuk parameter yang diubah, ada kernel, C, dan gamma. Kernel menjelaskan tipe pemisah hyperplane apa yang akan digunakan. C menjelaskan parameter penalti dari error. Bagian ini mengatur perbandingan antara pemisahan batas yang teratur dan pengelompokan data dengan benar. Terakhir, gamma menjelaskan parameter pengaturan untuk hyperplane selain linear, dimana nilai gamma yang makin tinggi, berarti fitting terhadap training data semakin dilakukan.

Sumber:

https://medium.com/all-things-ai/in-depth-parameter-tuning-for-svc-758215394769
https://medium.com/edureka/support-vector-machine-in-python-539dca55c26a

### SVC

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC

svm = SVC()
params = {
    "C" : [0.01, 0.1, 1, 10, 100],
    "kernel" : ["linear", "poly", "rbf"],
    "gamma" : [0.001, 0.01, 0.1, 1]
}

grid_svm_tfidf = GridSearchCV(svm, param_grid = params)
grid_svm_tfidf.fit(TFIDF_train, y_train)
```

```python
## Model Prediction
pred = grid_svm_tfidf.predict(TFIDF_test)
eval_svm_tfidf = {
    'accuracy' : accuracy_score(y_test, pred),
    'precision' : precision_score(y_test, pred, average="weighted"),
    'recall' : recall_score(y_test, pred, average="weighted"),
    'F1' : f1_score(y_test, pred, average="weighted")
}

print(classification_report(y_test, pred))
```

```python
svm_tfidf_bestparams = grid_svm_tfidf.best_params_
svm_tfidf_bestparams
```

```python
eval_svm_tfidf
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

**Random Forest** adalah salah satu model klasifikasi machine learning dimana dasarnya adalah sebuah kumpulan decision tree. Kumpulan decision tree ini dibuat dengan metode 'Bagging' yang bertujuan untuk mengurangi kompleksitas model yang bisa overfit terhadap training data. Sama halnya dengan SVC, ada beberapa parameter yang diubah yaitu **n_estimators** yang menjelaskan jumlah tree yang digunakan untuk membuat forest. **max_depth** menjelaskan tentang seberapa dalam pohonnya, berarti jaalur terpanjang dari root sampai ke daunnya. **Criterion** ini adalah function yang mengukur kualitas dari split.
